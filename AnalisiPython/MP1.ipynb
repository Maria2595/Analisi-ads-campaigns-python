```python
import pandas as pd
```

```python
campaign_data = pd.read_csv("D:/MP1/MP.csv")
```

```python
# Primo report con calcolo dei KPIs mancanti per avere un quandro più preciso sui risultati delle campagne da monitorare
campaign_report = (campaign_data
                   .assign(
                       ARPU = lambda x: (x['Revenue_Generated']/x['Conversions']),
                       Cost_per_Click = lambda x: (x['Budget']/x['Clicks']),
                       ROAS = lambda x: (x['Revenue_Generated']/x['Budget']),
                       Profit_Margin = campaign_data['Revenue_Generated'] - campaign_data['Budget']
                   )
                   .groupby(['Campaign_ID','Subscription_Tier'])
                   .agg(
                       Profit_Margin = ('Profit_Margin','sum'),
                       ROAS = ('ROAS','mean'),
                       ARPU = ('ARPU','mean'),
                       Cost_per_Click = ('Cost_per_Click','mean'),
                       Units_Sold = ('Units_Sold','sum'),
                       Total_Cost = ('Budget','sum'),
                       Total_Conversions = ('Conversions','sum'),
                       Avg_Discount_Level = ('Discount_Level','mean'),
                       ROI = ('ROI','sum'),
                       Subscription_Length = ('Subscription_Length','sum')
                   )
                   .round(0)
                   .reset_index()
)
```

```python
campaign_report.head(5)
```

```python
# Raggruppamento con calcolo di valori massimi, medi, minimi per poter satbilire dei range e attuare classificazioni
data_overview = (campaign_report.groupby('Subscription_Tier')
                               .agg(
                                Max_sales = ('Units_Sold','max'),
                                Avg_sales = ('Units_Sold','mean'),
                                Min_sales = ('Units_Sold','min'),
                                Max_Conversions = ('Total_Conversions','max'),
                                Avg_Conversions = ('Total_Conversions','mean'),
                                Min_Conversions = ('Total_Conversions','min'),
                                Max_arpu = ('ARPU','max'),
                                Mean_arpu = ('ARPU','mean'),
                                Min_arpu = ('ARPU','min'),
                                Max_CPC = ('Cost_per_Click','max'),
                                Mean_CPC = ('Cost_per_Click','mean'),
                                Min_CPC = ('Cost_per_Click','min'),
                                max_roas = ('ROAS','max'),
                                avg_roas = ('ROAS','mean'),
                                min_roas = ('ROAS','min'),
                                max_profit = ('Profit_Margin','max'),
                                avg_profit = ('Profit_Margin','mean'),
                                min_profit = ('Profit_Margin','min')
                               )
                               .round(0)
                               .reset_index()
)
```

```python
data_overview
```

```python
def ROAS_range (ROAS, valoremedio = 5 , valorealto = 100):
    if ROAS <= valoremedio:
        return 'ROAS Basso'
    elif ROAS <= valorealto:
        return 'ROAS Medio'
    else :
        return 'ROAS Alto'
```

```python
def CPC_range (Cost_per_Click, valoremedio = 34 , valorealto = 2000):
    if Cost_per_Click <= valoremedio:
        return 'CPC Basso'
    elif Cost_per_Click <= valorealto:
        return 'CPC Medio'
    else :
        return 'CPC Alto'
```

```python
def ARPU_range (ARPU, valoremedio = 300 , valorealto = 4000):
    if ARPU <= valoremedio:
        return 'ARPU Basso'
    elif ARPU <= valorealto:
        return 'ARPU Medio'
    else :
        return 'ARPU Alto'
```

```python
def Conversions_range (Conversions, valoremedio = 450 , valorealto = 900):
    if Conversions <= valoremedio:
        return 'Conversioni Basse'
    elif Conversions <= valorealto:
        return 'Conversioni Medie'
    else :
        return 'Conversioni Alte'
```

```python
def Profit_range (Profit_Margin, valoremedio = 2400, valorealto = 9600):
    if Profit_Margin <= valoremedio:
        return 'Livello Profitto Basso'
    if Profit_Margin <= valorealto:
        return 'Livello Profitto Medio'
    else : 
        return 'Livello Profitto Alto'
```

```python
# Report finale con aggiunta delle classificazioni alle variabili calcolate nel report rpecedente
campaign_report2 = (campaign_report
                    .assign(
                        Livello_ROAS = campaign_report['ROAS'].apply(ROAS_range),
                        Livello_ARPU = campaign_report['ARPU'].apply(ARPU_range),
                        Livello_Conversions = campaign_report['Total_Conversions'].apply(Conversions_range),
                        Livello_Cost_per_Click = campaign_report['Cost_per_Click'].apply(CPC_range),
                        Livello_Profitto = campaign_report['Profit_Margin'].apply(Profit_range)
                    )
                    .reset_index()
)
```
